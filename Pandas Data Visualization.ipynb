# Pandas Built-in Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df1 = pd.read_csv('df1',index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

```python
# Quick Example of what can be done w/ these dataframes and columns
# if we want a histogram of all the values in the A column from df1
df1['A'].hist(bins=30)
```

```python
df1['A'].plot(kind='hist',bins=30)
```

```python
# We can also just call .plot.hist() for the same result
df1['A'].plot.hist()
```

```python
df2.head()
```

```python
# Area Plots
df2.plot.area(alpha=0.4)
```

```python
# Bar Plots
df2.plot.bar()
```

```python
# We can also make this 'stacked' by passing in
df2.plot.bar(stacked=True)
```

```python
# Line Plots (for df1 = Time Series)
# for line plot, need to specify x and y
df1.plot.line(x=df1.index,y='B',figsize=(12,3),lw=1)
```

```python
# Scatter Plots
# c='C' inicates the color in comparison to another column. in this case its column C in df1
df1.plot.scatter(x='A',y='B',c='C',cmap='rainbow')
```

```python
df1.head()
```

```python
# Scatter Plots
# instead of passing in c='C' for color,
# we can do size by passing in s=dataframe['Column']*(factor)
df1.plot.scatter(x='A',y='B',s=df1['C']*100)
```

```python
# Boxplots
df2.plot.box()
```

```python
# Variant Data: Hexagonal Bin Plot

# we will make a new data frame for this:
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
df.head()
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25)
```

```python
# Kernal Density Estimation (KDE) Plots
df2['a'].plot.kde()
```

```python
# can also do it to the entire Data Frame as well
df2.plot.kde()
```
